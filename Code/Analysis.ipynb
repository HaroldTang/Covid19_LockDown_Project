```python
import pandas as pd
import numpy as np
import glob,os
from datetime import datetime
import plotly.graph_objects as go
import plotly.express as px
```

```python
# read apple mobility data
apple_df= pd.read_csv('../Output/Mobility_Result/apple_mobility.csv')
apple_df.head(3)
```

```python
# decomposite the dataset
apple_mobility = apple_df.melt(id_vars=['geo_type','region', 'transportation_type'],
                               var_name='date',value_name = 'mobility %')
apple_mobility.head(10)
```

```python
# For cities compare
fig = px.line(apple_mobility, x= 'date', y = 'mobility %', color="region",
              line_group="transportation_type", title='Apple mobility by city')
fig.update_xaxes(
    rangeselector=dict(
        buttons=list([
            dict(count=7, label="7d", step="day", stepmode="backward"),
            dict(count=14, label="2w", step="day", stepmode="backward"),
            dict(count=1, label="1m", step="month", stepmode="backward"),
            dict(count=2, label="2m", step="month", stepmode="backward"),
            dict(step="all")
        ])
    )
)
fig.show()
```

```python
# Transportation methods compare
Overall = apple_mobility[apple_mobility['geo_type'] == 'AllCountry']
fig = px.line(Overall, x= 'date', y= 'mobility %', 
              color="transportation_type",title='Apple mobility United States')
fig.show()
```

```python
transport_city = apple_mobility.drop(Overall.index)
fig = px.line(transport_city, x= 'date', y = 'mobility %',
              color="transportation_type",line_group="region", title='Apple mobility by transportation type')
fig.update_xaxes(
    rangeselector=dict(
        buttons=list([
            dict(count=7, label="7d", step="day", stepmode="backward"),
            dict(count=14, label="2w", step="day", stepmode="backward"),
            dict(count=1, label="1m", step="month", stepmode="backward"),
            dict(count=2, label="2m", step="month", stepmode="backward"),
            dict(step="all")
        ])
    )
)
fig.show()
```

```python
import plotly.figure_factory as ff
mobility_state= pd.read_csv('../Output/Mobility_Result/DL_mobility_state.csv')
m50_index_state = mobility_state[mobility_state['class'] == 'm50_index']
m50_index_state.head(10)
```

```python
mobility_county= pd.read_csv('../Output/Mobility_Result/DL_mobility_county.csv')
m50_index_county = mobility_county[mobility_county['class'] == 'm50_index']
m50_index_county
```

```python
values = m50_index_county['03-01'].tolist()
fips = m50_index_county['fips'].tolist()
colorscale = ["#171c42","#223f78","#1267b2","#4590c4","#8cb5c9","#b6bed5","#dab2be",
              "#d79d8b","#c46852","#a63329","#701b20","#3c0911"]

endpts = list(np.linspace(-75, 75, len(colorscale) - 1))
fig = ff.create_choropleth(fips = fips, values = values, colorscale = colorscale, show_state_data = True,
                           binning_endpoints = endpts, county_outline={'color':'rgb(255,255,255)', 'width':0.5},
                           lengend_title='% change', title='% change in county mobility')
```

```python
stock_df= pd.read_csv('../Output/Stock_Price_Result/Stock_Price_innerjoin_clean.csv')
stock_df
```

```python
date = pd.DataFrame(stock_df['Date'])
stock = stock_df.drop('Date', axis=1)
result = 100.0 * stock/stock.iloc[0]
stock_percent = pd.concat([date, result],axis=1)
stock_percent
```

```python
stock = stock_percent.melt(id_vars=['Date'], var_name='Company',value_name = 'Stock Percent %')

fig = px.line(stock, x='Date', y='Stock Percent %', range_x=['2019-11-26', '2020-04-24'],
                 title="Stock Price without weekend gaps", color = 'Company')
fig.update_xaxes(rangeslider_visible=True)
fig.show()
```
