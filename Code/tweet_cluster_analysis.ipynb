```python
import pandas as pd
import numpy as np
import glob,os
from functools import reduce

df = pd.read_csv('../Output/Tweets_Result/tweets_clean_replace_cluster.csv')
#df_final = reduce(lambda left,right: (left,right,on='gram', how='outer'), dl)
df = df.groupby('gram').sum().reset_index()
df_sum = df
df_sum['sum'] = df.apply(lambda x: x[1:].sum(), axis = 1)
df_sum.sort_values(by='sum',ascending=False,inplace = True)

#df_sum.to_csv("../Output/Tweets_Result/tweets_clean_replace_cluster_1.csv")
df_lines = df_sum[:16]
df_lines.reset_index(drop=True,inplace=True) 

df_final = df_lines
for index,row in df_lines.iterrows():
    #print(row['ADMIN_LEVEL_NAME']) 
    cur = str(row['gram'])
    cur = cur.replace('~','covid-19')
#     print(cur)
    df_final.loc[index,'gram'] = str(cur)
# manual adjust
df_final.loc[3,'gram'] = 'pandemia covid-19'
df_final.loc[12,'gram'] = 'dies covid-19'

df_final = df_lines.groupby('gram').sum()

# # df_lines.loc[3,'gram'] = 'pandemia covid-19'
# # df_lines = df_lines.groupby('gram').sum().reset_index()

df_final.sort_values(by='sum',ascending=False,inplace = True)
df_final.reset_index(inplace=True)

# df_final.to_csv("../Output/Tweets_Result/tweets_clean_replace_cluster_top.csv")
df_final
```

```python
# get sum of tweets of each gram

sum_number = df_final.loc[:,['gram','sum']]
sum_number.to_csv("../Output/Tweets_Result/tweets_top_sum.csv")
sum_number
```
