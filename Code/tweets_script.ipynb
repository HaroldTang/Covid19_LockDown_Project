```python
import pandas as pd
import numpy as np
import glob,os
from functools import reduce

path=r'../Raw_data/Tweets/'
file=glob.glob(os.path.join(path, "*.csv"))

dl= []
col_name=['gram']

for f in file:
    dfi = pd.read_csv(f)
    dl.append(dfi)
    s = f[24:].split("_")
    col_name.append(s[0])


df_final = reduce(lambda left,right: pd.merge(left,right,on='gram', how='outer'), dl)
df_final.columns = col_name
df_final.set_index('gram',inplace=True)
df_final.to_csv("../Output/Tweets_Result/tweets_clean.csv")
df_final


```

```python
# get everyday tweets sum

date_sum = df_final.apply(lambda x: x.sum())
date_sum = date_sum.to_frame()

date_sum.index.names = ['Date']
date_sum.columns = ['Sum']
date_sum.sort_index(inplace=True)
date_sum.to_csv("../Output/Tweets_Result/tweets_daily_sum.csv")
date_sum
```
