```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
google_country_grocery= pd.read_csv('../Raw_data/Mobility_Change/US/mobility-grocery-and-pharmacy.csv')
google_country_parks= pd.read_csv('../Raw_data/Mobility_Change/US/mobility-parks.csv')
google_country_workplaces= pd.read_csv('../Raw_data/Mobility_Change/US/mobility-workplaces.csv')
google_country_retail= pd.read_csv('../Raw_data/Mobility_Change/US/mobility-retail-and-recreation.csv')
google_country_residential= pd.read_csv('../Raw_data/Mobility_Change/US/mobility-residential.csv')
google_country_transit= pd.read_csv('../Raw_data/Mobility_Change/US/mobility-transit-stations.csv')
google_country_parks.head(10)
```

```python
frames = [google_country_grocery,google_country_parks,google_country_workplaces,google_country_retail,google_country_residential,google_country_transit]
for df_ in frames[1:]:
    google_country_grocery = google_country_grocery.merge(df_, on='date')
google_country = google_country_grocery.sort_values(by='date')
google_country_mobility = google_country.melt(id_vars=['date'],
                               var_name='category',value_name = 'mobility %')
google_country_mobility
```

```python
import plotly.graph_objects as go
import plotly.express as px
```

```python
fig = px.line(google_country_mobility, x= 'date', y = 'mobility %', color="category",
              title='Google mobility whole country')
fig.update_xaxes(
    rangeselector=dict(
        buttons=list([
            dict(count=7, label="7d", step="day", stepmode="backward"),
            dict(count=14, label="2w", step="day", stepmode="backward"),
            dict(count=1, label="1m", step="month", stepmode="backward"),
            dict(step="all")
        ])
    )
)
fig.show()
```

```python
import glob,os
from functools import reduce
path=r'../Raw_data/Mobility_Change/US/'
file=glob.glob(os.path.join(path, "**/*.csv"))
google_state= []
col_name = ['date']
for f in file:
    dfi = pd.read_csv(f)
    google_state.append(dfi)
    
    state_name = f[31:33]
    category = f[43:].split('.')[0]
    column = state_name+'-'+category
    col_name.append(column)
df_final = reduce(lambda left,right: pd.merge(left,right,on='date', how='outer'), google_state)
df_final.columns = col_name
df_final.set_index('date',inplace=True)
df_final
```

```python
residential=df_final[df_final.columns[df_final.columns.str.endswith('residential')]]
residential.reset_index(inplace = True)
residential = residential.sort_values('date')
residential = residential.melt(id_vars=['date'],
                               var_name='category',value_name = 'mobility %')
```

```python
fig = px.line(residential, x= 'date', y = 'mobility %', color="category", title='Apple mobility by city')
fig.update_xaxes(
    rangeselector=dict(
        buttons=list([
            dict(count=7, label="7d", step="day", stepmode="backward"),
            dict(count=14, label="2w", step="day", stepmode="backward"),
            dict(count=1, label="1m", step="month", stepmode="backward"),
            dict(count=2, label="2m", step="month", stepmode="backward"),
            dict(step="all")
        ])
    )
)
fig.show()
```
