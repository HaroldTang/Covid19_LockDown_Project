```python
import pandas as pd
import numpy as np
import glob,os
from functools import reduce

df = pd.read_csv('../Raw_data/Mobility_Change/covid-19_goverment_measures.csv')
to_drop = ['ISO','LOG_TYPE','TARGETED_POP_GROUP','NON_COMPLIANCE','SOURCE','SOURCE_TYPE']
df.drop(to_drop,inplace=True,axis = 1)
df.drop_duplicates(inplace=True)

#deal with 'NAN' ADMIN_LEVEL_NAME
admin_name = 'State'

for index,row in df.iterrows():
    #print(row['ADMIN_LEVEL_NAME']) 
    cur = row['ADMIN_LEVEL_NAME']
    if cur is np.nan:
        df.loc[index,'ADMIN_LEVEL_NAME'] = admin_name

    else:
        admin_name = cur

df = df[ ~ df['CATEGORY'].str.contains('Public health measures') ]
df = df[ ~ df['CATEGORY'].str.contains('Governance and socio-economic measures') ]

# index again 
df.reset_index(drop=True,inplace=True) 
# save file
df.to_csv("../Output/Mobility_Result/covid-19_goverment_measures_clean.csv")
df
```
