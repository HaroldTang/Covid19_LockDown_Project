```python
import pandas as pd
import numpy as np
import glob,os
from datetime import datetime
import plotly.graph_objects as go
import plotly.express as px
```

```python
stock_df= pd.read_csv('../Raw_data/Stock_Price/Stock_Price_Result/rate/all_rate_innerjoin.csv')
stock_df
```

```python
stock = stock_df.melt(id_vars=['Date'], var_name='Company',value_name = 'Stock Percent %')
stock['Stock Percent %'] = stock['Stock Percent %'] - 100
fig = px.line(stock, x='Date', y='Stock Percent %', range_x=['2019-11-26', '2020-04-24'],
                 title="Stock Price without weekend gaps", color = 'Company')
fig.show()
```

```python
import glob,os
from functools import reduce
path=r'../Raw_data/Stock_Price/Stock_Price_Result/rate'
file=glob.glob(os.path.join(path, "*.csv"))
fig_list = []
stock_list = []
for f in file:
    dfi = pd.read_csv(f)
    category = f.split('/')[-1].split('_',1)[0]
    if category == 'all':
        continue
    avg_df = dfi.set_index('Date')
    avg_df['average'] = avg_df.mean(1)
    avg_df['category']= category
    stock_list.append(avg_df[['average','category']])
    stock = dfi.melt(id_vars=['Date'], var_name='Company',value_name = 'Stock Percent %')
    stock['Stock Percent %'] = stock['Stock Percent %'] - 100
    fig_list.append( px.line(stock, x='Date', y='Stock Percent %', range_x=['2019-11-26', '2020-04-24'],
                     title="Stock Price Percent in %s area" % category, color = 'Company'))
```
