```python
import pandas as pd
import numpy as np
import glob,os
from functools import reduce

path=r'../Raw_data/Stock_Price/'
file=glob.glob(os.path.join(path, "*.csv"))

dl= []
col_name = ['Date']
for f in file:
    dfi = pd.read_csv(f)
    to_drop = ['Open','High','Low','Adj Close','Volume']
    dfi.drop(to_drop,inplace=True,axis = 1)
    dl.append(dfi)
    
    s = f[24:].split('.')[0]
    col_name.append(s)
    
df_final = reduce(lambda left,right: pd.merge(left,right,on='Date', how='outer'), dl)
df_final.columns = col_name
df_final.set_index('Date',inplace=True)
df_final.to_csv("../Output/Stock_Price_Result/Stock_Price_outerjoin_clean.csv")
df_final
```

```python
import pandas as pd
import numpy as np
import glob,os
from functools import reduce

path=r'../Raw_data/Stock_Price/'
file=glob.glob(os.path.join(path, "*.csv"))

dl= []
col_name = ['Date']
for f in file:
    dfi = pd.read_csv(f)
    to_drop = ['Open','High','Low','Adj Close','Volume']
    dfi.drop(to_drop,inplace=True,axis = 1)
    dfi.dropna(axis=0,how='any',inplace=True)
    dl.append(dfi)
    
    s = f[24:].split('.')[0]
    col_name.append(s)
    
df_final = reduce(lambda left,right: pd.merge(left,right,on='Date', how='inner'), dl)
df_final.columns = col_name
df_final.set_index('Date',inplace=True)
df_final.to_csv("../Output/Stock_Price_Result/Stock_Price_innerjoin_clean.csv")
df_final
```
